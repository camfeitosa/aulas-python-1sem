```python
#5. Crie um programa para calcular a média ponderada de 4 valores de vendas com entrada de dados pelo usuário. 
# Em seguida, exiba a porcentagem de cada valor de venda em relação à média ponderada.

valor1 = float(input("Insira o valor:"))
valor2 = float(input("Insira o valor:"))
valor3 = float(input("Insira o valor:"))
valor4 = float(input("Insira o valor:"))

peso1 = float(input("Peso da venda 1: "))
peso2 = float(input("Peso da venda 2: "))
peso3 = float(input("Peso da venda 3: "))
peso4 = float(input("Peso da venda 4: "))

media_ponderada = (valor1 * peso1 + valor2 * peso2 + valor3 * peso3 + valor4 * peso4) / (peso1 + peso2 + peso3 + peso4)

porcentagem_valor1 = (valor1 * peso1 / media_ponderada) * 100
porcentagem_valor2 = (valor2 * peso2 / media_ponderada) * 100
porcentagem_valor3 = (valor3 * peso3 / media_ponderada) * 100
porcentagem_valor4 = (valor4 * peso4 / media_ponderada) * 100

# Exibição dos resultados
print(f"Média Ponderada: {media_ponderada}")
print("Porcentagem dos valores em relação à média ponderada: ")
print(f"Valor 1: {porcentagem_valor1}%")
print(f"Valor 2: {porcentagem_valor2}%")
print(f"Valor 3: {porcentagem_valor3}%")
print(f"Valor 4: {porcentagem_valor4}%")
```
